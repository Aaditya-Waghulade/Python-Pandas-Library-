```python
import pandas as pd 
df = pd.read_csv("comon file.csv",parse_dates=['DAY'])
type(df.DAY)
df
```

```python
#Filling NaN Data with another data by Different Method
```

```python
# 1) fillna() it will simply replace Empty cells with 0
new_df=df.fillna(0)
```

```python
#what if you have to Fill Space with conditions
new_df = df.fillna({
    'TEMPERATURE' :"Not Mentioned",
    'WINSPEED':"Not Detected",
    'EVENT' : "No Event",
})
new_df
```

```python
new_df=df.fillna(method="ffill")#If there is any empty value then it will enter the privious value into the next empty cell
new_df
```

```python
new_df=df.fillna(method="ffill" ,limit=1)#It will have the limit to copy the privious data only once
new_df
```

```python
#To Detect Random values which are missed then the method called interpolate() is used to detect inbetween values
new_df=df.interpolate() #It calculate values between upper cell and down cell
new_df
```

```python
#It will drop all those values which doesnt contain any values
new_df = df.dropna()
new_df
```

```python
#It will drop whole data which doesnt contain any value
new_df = df.dropna(how="all")
new_df
```
