```python
import pandas as pd
import numpy as np
df = pd.read_csv("replace-functions.csv")
df
```

```python
new_df = df.replace(-9999,np.NaN)#1) Replacing the value with NaN
new_df
```

```python
# 2) if you want to replace two different values throught the help of dictionary
new_df = df.replace({
    'temperature' :-9999,
    'windspeed': -9999,
    'events' : '0'#As we know "0" is in the columns of Strings inputs
},np.NaN)
new_df
```

```python
# 3) With the help of List we can change data of two different cells 
new_df = df.replace([-9999,-8888],np.NaN)
new_df
```
