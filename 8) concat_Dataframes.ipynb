```python
#Concatinating Means When you want to join two aur more dataframes with each other
import pandas as pd
india_weather = pd.DataFrame({
    'city' : ["Savda","Bhusawal","Faizpur","Jalgaon"],
    'temp' : [25,30,28,29],
    'humidity' : [60,70,80,90],
})
india_weather
```

```python
newyork_weather = pd.DataFrame({
    'city' : ["Manhatan","DownTown","Market","IT Parks"],
    'temp' : [30,20,12,39],
    'humidity' : [10,30,40,60],
})
newyork_weather
```

```python
combined_data=pd.concat([india_weather,newyork_weather])
combined_data
```

```python
df=pd.concat([india_weather,newyork_weather],ignore_index=True)#It will Ignore the Index and will start from zero
df
```

```python
df=pd.concat([india_weather,newyork_weather],keys=["india","newyork"])#Keys will show that from where the actual data is started and stoppes
df
```

#Now if you want to get all data of special key then you can use loc
combined_data.loc["india"]

```python
df.loc["india"] 
```

```python
# 2) Example 2 
windspeed_df = pd.DataFrame({
    "city" : ["Delhi","Mumbai"],
    "windspeed" : [20,30]
},index=[0,1])
windspeed_df
```

```python
temp_df = pd.DataFrame({
    "city" : ["Mumbai","Delhi"],
    "temperature" : [45,40]
},index=[0,1])
temp_df
```

```python
pd.concat([temp_df,windspeed_df],axis=1)#It will create another column because we know that it have same citie and so just we have to add column of Windspeed in our frame
```
